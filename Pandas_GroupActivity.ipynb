# Pandas Group Activity

```python
import pandas as pd
```

```python
d={'Mouse':[150,200,300,400],
'Keyboard':[180,200,190,300],
'Scanner':[200,280,330,450]}
```

```python
df=pd.DataFrame(d,index=['Jan','Feb','March','April'])
```

## A. Write code to access data of Mouse and Scanner columns.

```python
#Bracket notation
df[ ['Mouse' , 'Scanner'] ]
```

```python
#.iloc
df.iloc[:,[0,2]]
```

```python
#.loc
df.loc[:,['Mouse','Scanner']]
```

### .loc/.iloc methods

## D. Write code to access data of all columns where mouse data is more than 200.

### Bracket notation

```python
df[df['Mouse']>200]
```

```python
#square bracket after df statemnt to display specific column
df[df['Mouse']>200][['Mouse']]
```

### .loc notation

```python
df[df.loc[:,'Mouse']>200]
```

### .iloc notation

```python
df[df.iloc[:,0]>200]
```

### Dot notation

```python
df.Mouse > 200
```

```python
df[df.Mouse > 200]
```

## F. Write code to access data of rows of jan and march for scanner and keyboard

```python
#integer indexing
df.iloc[[0,2],[2,1]]
```

```python
#label indexing
df.loc[['Jan','March']][['Scanner','Keyboard']]
```

```python
#Bracket notation
#df[[<col_1>,<col_2>]][<Row_1>:<Row_n>:<Step>]
df[['Scanner','Keyboard']]['Jan':'March':2]
```
